```python
import matplotlib.pyplot as plt
from collections import defaultdict

data = [
    (16, 42.1), (16, 41.8), (16, 43.0), (16, 45.8), (16, 39.0),
    (32, 38.7), (32, 39.1), (32, 37.9), (32, 43.1), (32, 42.9),
    (64, 35.2), (64, 34.8), (64, 35.5), (64, 29.8), (64, 30.5),
    (128, 33.8), (128, 34.1), (128, 31.1),
    (256, 34.5), (256, 34.2), (256, 37.2),
    (512, 36.9), (512, 37.3), (512, 41.3),
]

hidden_dims, mae = zip(*data)

plt.figure(figsize=(15,6))
plt.scatter(hidden_dims, mae, label="single test", color="blue", alpha=0.75) #<-- data points

bucket = defaultdict(list)
for h, m in data:
    bucket[h].append(m)

mean_hidden = sorted(bucket.keys())
mean_mae = [sum(bucket[h]) / len(bucket[h]) for h in mean_hidden]

plt.plot(mean_hidden, mean_mae, label="mean performance", marker='x', color="darkorange", linestyle=':') #<-- average line

plt.xscale("log")
plt.xlabel("Hidden Dimensions")
plt.ylabel("MAE")
#plt.xticks([16, 32, 64, 128, 256, 512], labels=[16, 32, 64, 128, 256, 512])
plt.xticks(16, 32, 64, 128, 256, 512)
plt.title("MAE based on hidden dimensions")
plt.legend() # <-- 'label' above assigns the legend
plt.grid(True)
plt.show()
```
