```python
import matplotlib.pyplot as plt
from collections import defaultdict
data = [
    (1, 27.07), (1, 28.35), (1, 27.66), (1, 27.13), (1, 27.38),
    (2, 25.84), (2, 26.34), (2, 26.62), (2, 26.14), (2, 26.6),
    (3, 27.93), (3, 28.72), (3, 28.08), (3, 27.69), (3, 27.4),
    (4, 28.83), (4, 28.53), (4, 29.13), (4, 28.75), (4, 28.71), 
]
"""
data = [
    (32, 27.07), (32, 28.35), (32, 27.66), (32, 27.13), (32, 27.38),
    (64, 25.84), (64, 26.34), (64, 26.62), (64, 26.14), (64, 26.6),
    (128, 27.93), (128, 28.72), (128, 28.08), (128, 27.69), (128, 27.4),
    (256, 28.83), (256, 28.53), (256, 29.13), (256, 28.75), (256, 28.71), 
]
"""
hidden_dims, mae = zip(*data)

plt.figure(figsize=(10,6))
plt.scatter(hidden_dims, mae, label="single test", color="blue", alpha=0.75) #<-- data points

bucket = defaultdict(list)
for h, m in data:
    bucket[h].append(m)

mean_hidden = sorted(bucket.keys())
mean_mae = [sum(bucket[h]) / len(bucket[h]) for h in mean_hidden]

plt.plot(mean_hidden, mean_mae, label="mean performance", marker='x', color="darkorange", linestyle=':') #<-- average line

#plt.xscale("log")
plt.xlabel("Layers (k-hop neighborhood)", fontsize=14)
plt.ylabel("MAE", fontsize=14)
plt.xticks([1, 2, 3, 4], labels=[1, 2, 3, 4], fontsize=14)
plt.yticks(fontsize=14)
plt.title("Overall MAE based on the number of layers", fontsize=18)
plt.legend(loc='lower right', fontsize=12) # <-- 'label' above assigns the legend
plt.grid(True)
plt.show()
```
