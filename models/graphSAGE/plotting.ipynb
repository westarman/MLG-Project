```python
import matplotlib.pyplot as plt
from collections import defaultdict

data = [
    (16, 42.1), (16, 41.8), (16, 43.0),
    (32, 38.7), (32, 39.1), (32, 37.9),
    (64, 35.2), (64, 34.8), (64, 35.5),
    (128, 33.8), (128, 34.1),
    (256, 34.5), (256, 34.2),
    (512, 36.9), (512, 37.3),
]

hidden_dims, mae = zip(*data)

plt.figure(figsize=(15,9))
plt.scatter(hidden_dims, mae, alpha=0.9) #<-- data points

bucket = defaultdict(list)
for h, m in data:
    bucket[h].append(m)

mean_hidden = sorted(bucket.keys())
mean_mae = [sum(bucket[h]) / len(bucket[h]) for h in mean_hidden]

plt.plot(mean_hidden, mean_mae, marker='x',color="green", linestyle='--') #<-- average line

plt.xlabel("Hidden Dimensions")
plt.ylabel("MAE")
plt.xticks([16, 32, 64, 128, 256, 512])

plt.grid(True)
plt.show()
```
