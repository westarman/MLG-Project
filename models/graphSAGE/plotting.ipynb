```python
import matplotlib.pyplot as plt
from collections import defaultdict

data = [
    ("sum", 27.91), ("sum", 28.17), ("sum", 28.47), ("sum", 28.48), ("sum", 28.35),
    ("mean", 25.84), ("mean", 26.36), ("mean", 26.62), ("mean", 26.14), ("mean", 26.6),
    ("max", 27.82), ("max", 27.66), ("max", 28.27), ("max", 28.74), ("max", 26.98)
]

hidden_dims, mae = zip(*data)

plt.figure(figsize=(10,6))
plt.scatter(hidden_dims, mae, label="single test", color="blue", alpha=0.75) #<-- data points

bucket = defaultdict(list)
for h, m in data:
    bucket[h].append(m)

mean_hidden = sorted(bucket.keys())
mean_mae = [sum(bucket[h]) / len(bucket[h]) for h in mean_hidden]

plt.plot(mean_hidden, mean_mae, label="mean performance", marker='x', color="darkorange", linestyle=':') #<-- average line

#plt.xscale("log")
plt.xlabel("Aggregation function", fontsize=14)
plt.ylabel("MAE", fontsize=14)
plt.xticks(["sum", "mean", "max"], labels=["sum", "mean", "max"], fontsize=14)
plt.yticks(fontsize=14)
plt.title("Overall MAE based on Aggregation function", fontsize=18)
plt.legend(loc='lower right', fontsize=12) # <-- 'label' above assigns the legend
plt.grid(True)
plt.show()
```
